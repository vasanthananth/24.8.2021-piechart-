```python
import matplotlib.pyplot as plotter
wedgeLabels     = ("Data Science", "ML", "DL", "AI", "SAP", "IT-Others")
ngramPercent    = (10, 20, 20, 30, 11, 76)
figureObject, axesObject = plotter.subplots()
axesObject.pie(ngramPercent,

               labels       = wedgeLabels,

               shadow       = False,

               frame        = True,              

               startangle   = 120,

               autopct      = '%.1f%%',

               wedgeprops   = { 'linewidth' : 1,

                                'edgecolor' : "black" })

axesObject.axis('equal')
plotter.show()
```
